# EfficientNet Training

**Objective:** Train and evaluate EfficientNet-B0 model for deepfake detection.

**Research Traceability:**
- Research Objective: EfficientNet-based deepfake detection with compound scaling
- Methodology: Transfer learning with EfficientNet-B0 pretrained on ImageNet
- Implementation: notebooks/05_training_efficientnet.ipynb

```python
import os
import sys
sys.path.insert(0, os.path.abspath('..'))

import torch
import matplotlib.pyplot as plt
from pathlib import Path

from src.models.efficientnet import EfficientNetModel

plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
```

```python
# Initialize EfficientNet-B0
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = EfficientNetModel(
    num_classes=2,
    model_name='efficientnet_b0',
    pretrained=True,
    dropout_rate=0.3,
)
model = model.to(device)

from src.utils.helpers import count_parameters
params = count_parameters(model)
print(f"Model: EfficientNet-B0")
print(f"Total parameters: {params['total']:,}")
print(f"Trainable parameters: {params['trainable']:,}")
print(f"Device: {device}")
```

## Training Configuration

```python
config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 50,
    'weight_decay': 0.0001,
    'label_smoothing': 0.1,
    'early_stopping_patience': 10,
    'image_size': 224,  # EfficientNet default
}

print("EfficientNet Training Configuration:")
for k, v in config.items():
    print(f"  {k}: {v}")
```

## Model Architecture Summary

EfficientNet-B0 uses compound scaling to balance depth, width, and resolution:
- **Depth:** 18 layers (MBConv blocks)
- **Width:** 1.0x scaling factor
- **Resolution:** 224x224 input
- **Parameters:** ~5.3M (vs ~22.9M for XceptionNet)

# XceptionNet Training

**Objective:** Train and evaluate the XceptionNet model for deepfake detection.

**Research Traceability:**

```python
import os
import sys
sys.path.insert(0, os.path.abspath('..'))

import torch
import matplotlib.pyplot as plt
from pathlib import Path

from src.models.xception import XceptionNet
from src.config.settings import Settings

plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
```

```python
# Initialize model
settings = Settings()
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = XceptionNet(
    num_classes=2,
    pretrained=True,
    dropout_rate=0.5,
)
model = model.to(device)

# Print model summary
from src.utils.helpers import count_parameters
params = count_parameters(model)
print(f"Model: XceptionNet")
print(f"Total parameters: {params['total']:,}")
print(f"Trainable parameters: {params['trainable']:,}")
print(f"Device: {device}")
```

## Training Configuration

```python
# Training hyperparameters
config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 50,
    'weight_decay': 0.0001,
    'label_smoothing': 0.1,
    'early_stopping_patience': 10,
}

print("Training Configuration:")
for k, v in config.items():
    print(f"  {k}: {v}")
```

## Training Loop

Note: Actual training requires the dataset to be prepared.
Run `scripts/preprocess.sh` first, then use `scripts/train.sh` for full training.

```python
# Placeholder for training visualization
# After training, plot training curves here

def plot_training_curves(history_path: str):
    """Plot training and validation curves."""
    import json
    
    if not Path(history_path).exists():
        print("No training history found")
        return
    
    with open(history_path) as f:
        history = json.load(f)
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Loss curve
    axes[0].plot(history['train_loss'], label='Train')
    axes[0].plot(history['val_loss'], label='Validation')
    axes[0].set_title('Loss')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    
    # Accuracy curve
    axes[1].plot(history['train_acc'], label='Train')
    axes[1].plot(history['val_acc'], label='Validation')
    axes[1].set_title('Accuracy')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()
    
    plt.tight_layout()
    plt.savefig('../outputs/plots/xceptionnet_training.png', dpi=150)
    plt.show()

# Uncomment after training:
# plot_training_curves('../outputs/models/xceptionnet_history.json')
```
